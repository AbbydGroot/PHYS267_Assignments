```python
import math 
import numpy as np
import random
import matplotlib.pyplot as plt


##Abigail deGroot
##21012489 
##PHYS 267 Assignment 1


##Q2: Coin Flip Problem

## provided "human" dataset 
X = "TTHTHHHHTTHTHHTTHTHTHHHHTTHTTTH"

## provided "random" dataset
O = "THHTHHHTHTTTHTTHTHTHTHHTHTTHHHH"


## stolen code from Dr. Martin https://gist.github.com/jddmartin/c7d1780702dd942c06eeb693e4cbc45f 
def run_counter(s):
    """ counts runs within a sequence
    returns a list, where the ith element of the list is the
    # number of runs of length i. """

    run_counts = [0,] * len(s)
    p = s[0]
    n = 1
    for v in s[1:]:
        if v == p:
            n += 1
        else:
            run_counts[n] += 1
            p = v
            n = 1
            
    # flush last run:
    if n != 1:
        run_counts[n] += 1

    return run_counts

## *  *  *  *  * end of given information *  *  * 

def longest_run(s): 
    """takes longest run from run_counts """
    l = run_counter(s)
    i = len(l)-1
    while l[i] == 0: 
        i -= 1
        longest = i 
            
    return longest 



def switch_counter(s):
    """counts switches within a sequence
    taking a list and returning an integer value"""
    n = 0
    p = s[0]
    for v in s[0:]:
        if v != p:
            n += 1 
            p = v
    
    return n



## * * * 



x_rand = []
y_rand = []

def rand_data():
    n_trials = 2000
    n_bits = 32
    for i in range(n_trials):
        # generate random binary sequence:
        s = [int(random.random()+0.5) for _ in range(n_bits)]
        x_rand.append(longest_run(s))
        y_rand.append(switch_counter(s))
    return x_rand, y_rand


rand_data()



#provided points (a)
plt.plot(longest_run(list(X)), switch_counter(list(X)), 'rX', label = 'provided fake random' )
plt.plot(longest_run(list(O)), switch_counter(list(O)), 'ro', label = 'provided coin toss')

#fabricated random sequence  (b)
plt.plot(longest_run("HTHTHHTTHTHHHTHHTTHHHTHTHTTTTTHT"), switch_counter("HTHTHHTTHTHHHTHHTTHHHTHTHTTTTTHT"), 'gs', label ='fake random')
#coin toss
plt.plot(longest_run("HHHTHHHTHTTTHTHTTHTHTHHTHTTHHHT"), switch_counter("HHHTHHHTHTTTHTHTTHTHTHHTHTTHHHT"), 'gD', label = 'coin toss')

#random points 
plt.scatter(x_rand, y_rand, marker = '.', label = 'random points')
plt.xlabel("longest run in sequence")
plt.ylabel("number of switches in sequence")
plt.legend()

"""c) I think that if I were the instructor, I would be willing to take the risk to do the demo in class, but only if I 
   were doing it during the first week. If the demo succeeds it would be really cool for students to see statistics in action. 
   If the demo failed, this may make the students feel closer to you, and they can see how professionals handle 
   unexpected results. Either way the result is a learning opportunity for the students and builds repore because it is interactive."""
```
